```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
%matplotlib inline
```

```python
df = pd.read_excel("J0331.xlsx", index_col = 0)
# apre il file di excel, tramite index_col=0 gli dico che la colonna della data è la colonna 0 per cui lui non ne tiene conto per i conti e la usa come asse delle ascisse nel grafico
# Se non facessi il filtraggio della colonna data, lui comunque si accorgerebbe che è una colonna data per cui potrei gestire il maggiore o minore o altre cose, capendo le date
```

```python
mp.rcParams["figure.figsize"] = (30,15) # serve per attribuire una dimensione fissa al grafico
df[["T5", "T6"]].plot() # serve per plottare il grafico, visualizzando solo le colonne T5 e T6 in questo caso. Tra parentesi quadre gli passo la lista degli elementi che voglio
# visualizzare
df.head() # fa vedere le prime 5 colonne
```

```python
df.rename({"T1":"TempCamma"}, axis="columns", inplace=True )
# il dizionario è un mapper, gli passo quindi il nome attuale della colonna e dopo : quello che voglio dargli. Poi axis può essere "index" o "columns", indicando la colonna dei
# riferimenti (la pridfma colonna a sinistra) oppure quella dei campi (quella sopra le colonne). POi ultimo campo con inplace True gli dico che voglio sostituire il valore nella
# tabella del file excel che ho create (df) non creare un oggetto nuovo
```
